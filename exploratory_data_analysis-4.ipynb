# Company Ranking

```r
print('health')

head(health)
health_tick=as.vector(health$Ticker)

health_df=data.frame()
for (g in 1:length(health_tick)){
    temp=equity[equity$Ticker==health_tick[g],]
    health_df=rbind(health_df,temp)
}

head(health_df)

health_df$Currency=0
health_df$Name=0
print('bla')
head(health_df)
for (g in 1:length(health_tick)){
    health_df$Currency = as.character(health[health$Ticker==health_tick[g],]$Currency)
    health_df$Name = as.character(health[health$Ticker==health_tick[g],]$Description)
}

head(health_df)

dim(health_df)
dim(equity)
print('health_df1')
head(health_df)
health_df$Date=as.Date(health_df$Date)
# rank the company for every day
dates=as.vector(unique(health_df$Date))
print('health_df2')
str(health_df)
health_df=data.frame(Ticker=health_df$Ticker,Close=health_df$Close,Date=health_df$Date)

# replace NA with 0
health_df[is.na(health_df)]<-0
print('health_df3')
head(health_df)
empt=data.frame()
str(health_df)
for (d in 1:length(dates)){
    temp=health_df[health_df$Date==dates[d],] 
    temp=temp[order(temp$Close),]
    temp$rank = seq(1:dim(temp)[1])
    empt=rbind(empt,temp)
}
empt = empt[order(-empt$Close),]
empt$rank=seq(1:dim(empt)[1])
head(empt)
```
