```python
data = pd.read_csv("british_comp.csv")
```

```python
data["Date"] = pd.to_datetime(data["Date"], format='%d/%m/%Y')
```

# AstraZeneca|

```python
AstraZeneca = data[data.Ticker == 'AstraZeneca PLC']
AstraZeneca.sort("Date", ascending=False, inplace=True)

AstraZeneca = AstraZeneca[['Date', 'Close']]
AstraZeneca

for i in range(0, 5):
    temp = AstraZeneca[0:len(AstraZeneca.index) - 1 - i].copy()
    temp['previous' + str(i)] = list(AstraZeneca[i + 1:len(AstraZeneca.index)].Close)
    AstraZeneca = temp.copy()

AstraZeneca.to_csv("AstraZeneca_PLC.csv")
```
